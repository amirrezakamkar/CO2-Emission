```python
url = '/Users/amirrezakamkar/Desktop/Works_in_progress/Ironhack/final_project/Data/clean/fossil_CO2_by_sector_and_country_clean.csv'
df = pd.read_csv(url)
```

```python
df = df.drop(columns=['Substance', 'EDGAR Country Code'])
```

```python
#changing the shape of the dataframe.
df_melted = df.melt(id_vars=["Country", "Sector"], var_name="Year", value_name="CO2_Emissions")
```

```python
df_melted.info()
```

```python
#cconverting the year column to integer.
df_melted["Year"] = df_melted["Year"].astype(int)
```

```python
# comparing the co2 emission in power industry in each decade
power_df = df_melted[df_melted["Sector"] == "Power Industry"]

bins = [1970, 1980, 1990, 2000, 2010, 2021]  
labels = ["1970-1979", "1980-1989", "1990-1999", "2000-2009", "2010-2021"]
power_df["Decade"] = pd.cut(power_df["Year"], bins=bins, labels=labels, right=False)

decade_summary_power = power_df.groupby("Decade")["CO2_Emissions"].sum().reset_index()

decade_summary_power
```

```python
decade_summary_power["Percentage_Increase"] = decade_summary_power["CO2_Emissions"].pct_change() * 100

decade_summary_power
```

```python
decade_summary_power.to_csv("power_sector_decade_summary.csv", index=False)
```

```python
# comparing the co2 emission in transport industry in each decade
transport_df = df_melted[df_melted["Sector"] == "Transport"]

bins = [1970, 1980, 1990, 2000, 2010, 2021]  
labels = ["1970-1979", "1980-1989", "1990-1999", "2000-2009", "2010-2021"]
transport_df["Decade"] = pd.cut(transport_df["Year"], bins=bins, labels=labels, right=False)

decade_summary_transport = transport_df.groupby("Decade")["CO2_Emissions"].sum().reset_index()

decade_summary_transport["Percentage_Increase"] = decade_summary_transport["CO2_Emissions"].pct_change() * 100

decade_summary_transport
```

```python
decade_summary_transport.to_csv("transport_sector_decade_summary.csv", index=False)
```

```python
# 
top_10_economies = [
    "United States", "China", "Japan", "Germany", "United Kingdom",
    "India", "France", "Italy", "Canada", "South Korea"
]

df_2021_top_power = df_melted[
    (df_melted["Sector"] == "Power Industry") &
    (df_melted["Country"].isin(top_10_economies))
]

bins = [1970, 1980, 1990, 2000, 2010, 2021]
labels = ["1970-1979", "1980-1989", "1990-1999", "2000-2009", "2010-2021"]

df_2021_top_power = df_2021_top_power.assign(
    Decade=pd.cut(df_2021_top_power["Year"], bins=bins, labels=labels, right=False)
)

decade_top_power = df_2021_top_power.groupby("Decade")["CO2_Emissions"].sum().reset_index()
decade_top_power["Percentage_Increase"] = decade_top_power["CO2_Emissions"].pct_change() * 100

decade_top_power
```

```python
# 

df_usa = df_melted[
    (df_melted["Sector"] == "Power Industry") &
    (df_melted["Country"] == "United States")
]

bins = [1970, 1980, 1990, 2000, 2010, 2021]
labels = ["1970-1979", "1980-1989", "1990-1999", "2000-2009", "2010-2021"]

df_usa = df_usa.assign(
    Decade=pd.cut(df_usa["Year"], bins=bins, labels=labels, right=False)
)

usa_decade = df_usa.groupby("Decade")["CO2_Emissions"].sum().reset_index()
usa_decade["Percentage_Increase"] = usa_decade["CO2_Emissions"].pct_change() * 100

usa_decade
```

```python
usa_decade.to_csv("usa_decades.csv", index=False)
```

```python
df_china = df_melted[
    (df_melted["Sector"] == "Power Industry") &
    (df_melted["Country"] == "China")
]

bins = [1970, 1980, 1990, 2000, 2010, 2021]
labels = ["1970-1979", "1980-1989", "1990-1999", "2000-2009", "2010-2021"]

df_china = df_china.assign(
    Decade=pd.cut(df_china["Year"], bins=bins, labels=labels, right=False)
)

china_decade = df_china.groupby("Decade")["CO2_Emissions"].sum().reset_index()
china_decade["Percentage_Increase"] = china_decade["CO2_Emissions"].pct_change() * 100

china_decade
```

```python
china_decade.to_csv("china_decades.csv", index=False)
```

```python
# checking the possible decrease in co2 emission in power industry fo top 10 most environmentally friendly countries.
# source: https://worldpopulationreview.com/country-rankings/most-environmentally-friendly-countries
top_10_green = ["Estonia", "Luxembourg", "Germany", "Finland", "United Kingdom", 
                "Sweden", "Norway", "Austria", "Switzerland", "Denmark"
               ]
df_green_2010 = df_melted[
    (df_melted["Sector"] == "Power Industry") &
    (df_melted["Country"].isin(top_10_green)) &
    (df_melted["Year"] == 2010)
]
display(df_green_2010)
df_green_2010_emission = df_green_2010["CO2_Emissions"].sum()
print("The total co2 emission of top 10 most green country in Power Industry in 2010 is ", df_green_2010_emission)
```

```python
df_green_2021 = df_melted[
    (df_melted["Sector"] == "Power Industry") &
    (df_melted["Country"].isin(top_10_green)) &
    (df_melted["Year"] == 2021)
]
display(df_green_2021)
df_green_2021_emission = df_green_2021["CO2_Emissions"].sum()
print("The total co2 emission of top 10 most green country in Power Industry in 2021 is ", df_green_2021_emission)
```

```python
# Displaying the data in a dataframe, comparing the co2 emission of top 10 green countries in 2010 and 2021.
df_green = df_melted[
    (df_melted["Sector"] == "Power Industry") &
    (df_melted["Country"].isin(top_10_green)) &
    (df_melted["Year"].isin([2010, 2021]))
]

df_green_summary = (
    df_green.groupby("Year")["CO2_Emissions"]
    .sum()
    .reset_index()
    .rename(columns={"CO2_Emissions": "Total_CO2_Emissions"})
)

df_green_summary
```

```python
df_green_summary.to_csv("green_countries_2010_2021_comparison.csv", index=False)
```
