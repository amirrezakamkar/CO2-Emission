```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = '/Users/amirrezakamkar/Desktop/Works_in_progress/Ironhack/final_project/Data/clean/fossil_CO2_per_GDP_by_country_clean.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.Country.unique()
```

```python
#removing the rows which shows the co2 emission in total and for EU.
df = df[~df["Country"].isin(['EU27', 'GLOBAL TOTAL'])]
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Substance', 'EDGAR Country Code'])
```

```python
#changing the shape of the dataframe.
df_melted = df.melt(id_vars=["Country"], var_name="Year", value_name="CO2_Emissions")
```

```python
df_melted.head()
```

```python
df_melted.info()
```

```python
#cconverting the year column to integer.
df_melted["Year"] = df_melted["Year"].astype(int)
```

```python
df_2021 = df_melted[df_melted["Year"] == 2021]

df_top10_2021 = df_2021.sort_values(by="CO2_Emissions", ascending=False).head(11)

df_top10_2021
```

```python
df_top10_2021 = df_top10_2021[~df_top10_2021["Country"].isin(['Palau'])]
```

```python
df_top10_2021
```

```python
df_top10_2021.to_csv("top_10_co2_per_gdp_2021.csv", index=False)
```

```python
top_10_green = ["Estonia", "Luxembourg", "Germany", "Finland", "United Kingdom", 
                "Sweden", "Norway", "Austria", "Switzerland", "Denmark"
               ]
df_green_2021 = df_melted[
    (df_melted["Country"].isin(top_10_green)) &
    (df_melted["Year"] == 2021)
]
display(df_green_2021)
```

```python
df_green_2021.to_csv("top_10_green_co2_per_gdp_2021.csv", index=False)
```

```python
# comparing the co2 emission per gdp of the top 10 economies in 2021 and all the other countries.
top_10_economies = [
    "United States", "China", "Japan", "Germany", "United Kingdom",
    "India", "France and Monaco", "Italy, San Marino and the Holy See", "Canada", "South Korea"
]


df_top_10_economies = df_melted[
    (df_melted["Country"].isin(top_10_economies)) &
    (df_melted["Year"] == 2021)
]


df_top_10_economies_sorted = df_top_10_economies.sort_values(by="CO2_Emissions", ascending=False)

df_top_10_economies_sorted
```

```python
df_top_10_economies_sorted.to_csv("top_10_economies_co2_per_gdp_2021.csv", index=False)
```
